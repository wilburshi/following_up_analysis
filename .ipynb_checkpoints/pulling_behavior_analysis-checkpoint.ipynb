```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import string
import warnings
```

```python
# Smallest Difference between
# two Arrays
import sys
def findSmallestDifference(A, B, m, n):
    A.sort()
    B.sort()
    a = 0
    b = 0

    # Initialize result as max value
    result = sys.maxsize

    # Scan Both Arrays upto
    # sizeof of the Arrays
    while (a < m and b < n):
        if (abs(A[a] - B[b]) < result):
            result = abs(A[a] - B[b])

        # Move Smaller Value
        if (A[a] < B[b]):
            a += 1

        else:
            b += 1
    # return final sma result
    return result
```

# pulling behavior analysis
## load behavioral results

```python
# claim the list of sessions
animal1_name = "Dodson"
animal2_name = "Scorch"
self_session_days = ["20220920","20220922","20220923","20221010"]
mutual_session_days = ["20221011","20221013","20221014","20221015"]
n_self_sessions = np.shape(self_session_days)[0]
n_mutual_sessions = np.shape(mutual_session_days)[0]
```

```python
# load self session behavioral results
cumulative_pulls_scorch_self = {}
cumulative_pulls_dodson_self = {}

coop_pulls_num_1s_self = np.zeros([1,n_self_sessions])
coop_pulls_num_2s_self = np.zeros([1,n_self_sessions])
coop_pulls_num_3s_self = np.zeros([1,n_self_sessions])

for ii in np.arange(0,n_self_sessions,1):
    
    bhv_data_path = "/home/ws523/marmoset_tracking_bhv_data_from_task_code/"+self_session_days[ii]+"_"+animal1_name+"_"+animal2_name+"/"
    try:
        trial_record_json = bhv_data_path+self_session_days[ii]+"_Scorch_Dodson_TrialRecord_1.json"
        bhv_data_json = bhv_data_path+self_session_days[ii]+"_Scorch_Dodson_bhv_data_1.json"
        session_info_json = bhv_data_path+self_session_days[ii]+"_Scorch_Dodson_session_info_1.json"
        #
        trial_record = pd.read_json(trial_record_json)
        bhv_data = pd.read_json(bhv_data_json)
        session_info = pd.read_json(session_info_json)
    except:
        trial_record_json = bhv_data_path+self_session_days[ii]+"_Dodson_Scorch_TrialRecord_1.json"
        bhv_data_json = bhv_data_path+self_session_days[ii]+"_Dodson_Scorch_bhv_data_1.json"
        session_info_json = bhv_data_path+self_session_days[ii]+"_Dodson_Scorch_session_info_1.json"
        #
        trial_record = pd.read_json(trial_record_json)
        bhv_data = pd.read_json(bhv_data_json)
        session_info = pd.read_json(session_info_json)
        
    #
    animal1 = session_info['lever1_animal'][0].lower()
    animal2 = session_info['lever2_animal'][0].lower()
    #
    # clean up the trial_record
    warnings.filterwarnings('ignore')
    trial_record_clean = pd.DataFrame(columns=trial_record.columns)
    for itrial in np.arange(0,np.max(trial_record['trial_number']),1):
        # trial_record_clean.loc[itrial] = trial_record[trial_record['trial_number']==itrial+1].iloc[[0]]
        trial_record_clean = trial_record_clean.append(trial_record[trial_record['trial_number']==itrial+1].iloc[[0]])
    trial_record_clean = trial_record_clean.reset_index(drop = True)
    #
    # change bhv_data time to the absolute time
    time_points_new = pd.DataFrame(np.zeros(np.shape(bhv_data)[0]),columns=["time_points_new"])
    for itrial in np.arange(0,np.max(trial_record_clean['trial_number']),1):
        ind = bhv_data["trial_number"]==itrial+1
        new_time_itrial = bhv_data[ind]["time_points"] + trial_record_clean["trial_starttime"].iloc[itrial]
        time_points_new["time_points_new"][ind] = new_time_itrial
    ind = bhv_data["trial_number"]>np.max(trial_record_clean['trial_number'])
    bhv_data["time_points"] = time_points_new["time_points_new"]
    bhv_data = bhv_data.drop(np.where(ind)[0])
    
    animal1_pulltime = bhv_data["time_points"][bhv_data["behavior_events"]==1].reset_index(drop=True)
    animal2_pulltime = bhv_data["time_points"][bhv_data["behavior_events"]==2].reset_index(drop=True)
    
    if (animal1 == "scorch"):
        cumulative_pulls_scorch_self[self_session_days[ii]] = animal1_pulltime
        cumulative_pulls_dodson_self[self_session_days[ii]] = animal2_pulltime
    else:
        cumulative_pulls_scorch_self[self_session_days[ii]] = animal2_pulltime
        cumulative_pulls_dodson_self[self_session_days[ii]] = animal1_pulltime
        
        
    #    
    #    
    # count the cooperative pull numbers
    for itrial in np.arange(0,np.max(bhv_data['trial_number']),1):
        ind1 = (bhv_data["trial_number"]==itrial+1) & (bhv_data["behavior_events"]==1)
        itimepoint_1 = bhv_data[ind1]["time_points"].reset_index(drop=True)
        itimepoint_1 = pd.Series.to_list(itimepoint_1)
        ind2 = (bhv_data["trial_number"]==itrial+1) & (bhv_data["behavior_events"]==2)
        itimepoint_2 = bhv_data[ind2]["time_points"].reset_index(drop=True)
        itimepoint_2 = pd.Series.to_list(itimepoint_2)
        if ((len(itimepoint_1)>0) & (len(itimepoint_2)>0)):
            pull_diff = findSmallestDifference(itimepoint_1, itimepoint_2, len(itimepoint_1), len(itimepoint_2))
        if (pull_diff<=3):
            coop_pulls_num_3s_self[0][ii] = coop_pulls_num_3s_self[0][ii]+1
        if (pull_diff<=2):
            coop_pulls_num_2s_self[0][ii] = coop_pulls_num_2s_self[0][ii]+1
        if (pull_diff<=1):
            coop_pulls_num_1s_self[0][ii] = coop_pulls_num_1s_self[0][ii]+1
    coop_pulls_num_3s_self[0][ii] = coop_pulls_num_3s_self[0][ii]/(np.sum(bhv_data["behavior_events"]==1)+np.sum(bhv_data["behavior_events"]==2))
    coop_pulls_num_2s_self[0][ii] = coop_pulls_num_2s_self[0][ii]/(np.sum(bhv_data["behavior_events"]==1)+np.sum(bhv_data["behavior_events"]==2))
    coop_pulls_num_1s_self[0][ii] = coop_pulls_num_1s_self[0][ii]/(np.sum(bhv_data["behavior_events"]==1)+np.sum(bhv_data["behavior_events"]==2))
```

```python
# load mutual cooperation session behavioral results
cumulative_pulls_scorch_mutual = {}
cumulative_pulls_dodson_mutual = {}

coop_pulls_num_1s_mutual = np.zeros([1,n_mutual_sessions])
coop_pulls_num_2s_mutual = np.zeros([1,n_mutual_sessions])
coop_pulls_num_3s_mutual = np.zeros([1,n_mutual_sessions])

for ii in np.arange(0,n_mutual_sessions,1):
    
    bhv_data_path = "/home/ws523/marmoset_tracking_bhv_data_from_task_code/"+mutual_session_days[ii]+"_"+animal1_name+"_"+animal2_name+"/"
    try:
        trial_record_json = bhv_data_path+mutual_session_days[ii]+"_Scorch_Dodson_TrialRecord_1.json"
        bhv_data_json = bhv_data_path+mutual_session_days[ii]+"_Scorch_Dodson_bhv_data_1.json"
        session_info_json = bhv_data_path+mutual_session_days[ii]+"_Scorch_Dodson_session_info_1.json"
        #
        trial_record = pd.read_json(trial_record_json)
        bhv_data = pd.read_json(bhv_data_json)
        session_info = pd.read_json(session_info_json)
    except:
        trial_record_json = bhv_data_path+mutual_session_days[ii]+"_Dodson_Scorch_TrialRecord_1.json"
        bhv_data_json = bhv_data_path+mutual_session_days[ii]+"_Dodson_Scorch_bhv_data_1.json"
        session_info_json = bhv_data_path+mutual_session_days[ii]+"_Dodson_Scorch_session_info_1.json"
        #
        trial_record = pd.read_json(trial_record_json)
        bhv_data = pd.read_json(bhv_data_json)
        session_info = pd.read_json(session_info_json)
        
    #
    animal1 = session_info['lever1_animal'][0].lower()
    animal2 = session_info['lever2_animal'][0].lower()
    #
    # clean up the trial_record
    warnings.filterwarnings('ignore')
    trial_record_clean = pd.DataFrame(columns=trial_record.columns)
    for itrial in np.arange(0,np.max(trial_record['trial_number']),1):
        # trial_record_clean.loc[itrial] = trial_record[trial_record['trial_number']==itrial+1].iloc[[0]]
        trial_record_clean = trial_record_clean.append(trial_record[trial_record['trial_number']==itrial+1].iloc[[0]])
    trial_record_clean = trial_record_clean.reset_index(drop = True)
    #
    # change bhv_data time to the absolute time
    time_points_new = pd.DataFrame(np.zeros(np.shape(bhv_data)[0]),columns=["time_points_new"])
    for itrial in np.arange(0,np.max(trial_record_clean['trial_number']),1):
        ind = bhv_data["trial_number"]==itrial+1
        new_time_itrial = bhv_data[ind]["time_points"] + trial_record_clean["trial_starttime"].iloc[itrial]
        time_points_new["time_points_new"][ind] = new_time_itrial
    ind = bhv_data["trial_number"]>np.max(trial_record_clean['trial_number'])
    bhv_data["time_points"] = time_points_new["time_points_new"]
    bhv_data = bhv_data.drop(np.where(ind)[0])
       
    # save pulling time for cumulative pulls 
    animal1_pulltime = bhv_data["time_points"][bhv_data["behavior_events"]==1].reset_index(drop=True)
    animal2_pulltime = bhv_data["time_points"][bhv_data["behavior_events"]==2].reset_index(drop=True)
    
    if (animal1 == "scorch"):
        cumulative_pulls_scorch_mutual[mutual_session_days[ii]] = animal1_pulltime
        cumulative_pulls_dodson_mutual[mutual_session_days[ii]] = animal2_pulltime
    else:
        cumulative_pulls_scorch_mutual[mutual_session_days[ii]] = animal2_pulltime
        cumulative_pulls_dodson_mutual[mutual_session_days[ii]] = animal1_pulltime
        
    #    
    #    
    # count the cooperative pull numbers
    for itrial in np.arange(0,np.max(bhv_data['trial_number']),1):
        ind1 = (bhv_data["trial_number"]==itrial+1) & (bhv_data["behavior_events"]==1)
        itimepoint_1 = bhv_data[ind1]["time_points"].reset_index(drop=True)
        itimepoint_1 = pd.Series.to_list(itimepoint_1)
        ind2 = (bhv_data["trial_number"]==itrial+1) & (bhv_data["behavior_events"]==2)
        itimepoint_2 = bhv_data[ind2]["time_points"].reset_index(drop=True)
        itimepoint_2 = pd.Series.to_list(itimepoint_2)
        if ((len(itimepoint_1)>0) & (len(itimepoint_2)>0)):
            pull_diff = findSmallestDifference(itimepoint_1, itimepoint_2, len(itimepoint_1), len(itimepoint_2))
        if (pull_diff<=3):
            coop_pulls_num_3s_mutual[0][ii] = coop_pulls_num_3s_mutual[0][ii]+1
        if (pull_diff<=2):
            coop_pulls_num_2s_mutual[0][ii] = coop_pulls_num_2s_mutual[0][ii]+1
        if (pull_diff<=1):
            coop_pulls_num_1s_mutual[0][ii] = coop_pulls_num_1s_mutual[0][ii]+1
    coop_pulls_num_3s_mutual[0][ii] = coop_pulls_num_3s_mutual[0][ii]/(np.sum(bhv_data["behavior_events"]==1)+np.sum(bhv_data["behavior_events"]==2))
    coop_pulls_num_2s_mutual[0][ii] = coop_pulls_num_2s_mutual[0][ii]/(np.sum(bhv_data["behavior_events"]==1)+np.sum(bhv_data["behavior_events"]==2))
    coop_pulls_num_1s_mutual[0][ii] = coop_pulls_num_1s_mutual[0][ii]/(np.sum(bhv_data["behavior_events"]==1)+np.sum(bhv_data["behavior_events"]==2))
```
