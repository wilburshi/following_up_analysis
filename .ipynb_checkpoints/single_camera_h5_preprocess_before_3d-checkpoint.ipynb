```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
# import deeplabcut
```

```python
# load h5 files for camera1 and camera2
camera12_analyzed_path = "/ysm-gpfs/pi/jadi/VideoTracker_SocialInter/test_video_3d/20220920_Dodson_Scorch_camera12/"
camera1_h5_file = camera12_analyzed_path + "20220920_Dodson_Scorch_camera-1DLC_dlcrnetms5_marmoset_tracking_with_middle_cameraSep1shuffle1_150000_el_filtered_old.h5"
camera2_h5_file = camera12_analyzed_path + "20220920_Dodson_Scorch_camera-2DLC_dlcrnetms5_marmoset_tracking_with_middle_cameraSep1shuffle1_150000_el_filtered_old.h5"
camera1_h5_file_save = camera12_analyzed_path + "20220920_Dodson_Scorch_camera-1DLC_dlcrnetms5_marmoset_tracking_with_middle_cameraSep1shuffle1_150000_el_filtered.h5"
camera2_h5_file_save = camera12_analyzed_path + "20220920_Dodson_Scorch_camera-2DLC_dlcrnetms5_marmoset_tracking_with_middle_cameraSep1shuffle1_150000_el_filtered.h5"
pd.read_hdf(camera2_h5_file)
```

```python
# modify camera1 and camera2 to have the same frame number
camera1_h5_data = pd.read_hdf(camera1_h5_file)
nframes_cam1_h5 = np.shape(camera1_h5_data)[0]
camera2_h5_data = pd.read_hdf(camera2_h5_file)
nframes_cam2_h5 = np.shape(camera2_h5_data)[0]
nframes_shared_h5 = np.min([nframes_cam1_h5,nframes_cam2_h5])
camera1_h5_data = camera1_h5_data.iloc[np.arange(0,nframes_shared_h5,1),:]
camera2_h5_data = camera2_h5_data.iloc[np.arange(0,nframes_shared_h5,1),:]
# save it to the new h5 files
camera1_h5_data.to_hdf(camera1_h5_file_save, key = "camera1_h5_data")
camera2_h5_data.to_hdf(camera2_h5_file_save, key = "camera2_h5_data")
```
